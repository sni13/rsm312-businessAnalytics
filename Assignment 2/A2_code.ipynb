### Transforming Data in original Files

```python
# Converting the Amount column to float
data_2015['Amount'] = data_2015['Amount'].str.replace('[\$,)]', '', regex=True)

# Changing (5.15) to -5.15
data_2015['Amount'] = data_2015['Amount'].str.replace('(', '-', regex = True)

data_2015['Amount'] = data_2015['Amount'].astype(float)
```

```python
data_2015.dtypes
```

```python
# Removing WhiteSpaces from the column names before merging -> as per instructions

# remove spaces from column names of data_2014
data_2014.rename(columns=lambda x: x.replace(' ', ''), inplace=True)

# remove spaces from column names of data_2015
data_2015.rename(columns=lambda x: x.replace(' ', ''), inplace=True)
```

```python
dataset = pd.concat([data_2014, data_2015]).reset_index(drop=True)
```

```python
dataset.info()
```

```python
dataset['TransactionDate'] = pd.to_datetime(dataset['TransactionDate'], format='%m/%d/%Y %H:%M')
dataset['PostedDate'] = pd.to_datetime(dataset['PostedDate'], format='%m/%d/%Y %H:%M')

# All times are 0:00, remove all of them
# Not sure if I should set NaT's to ''
```

```python
dataset['MerchantCategoryCode(MCC)'] = dataset['MerchantCategoryCode(MCC)'].str.lower()
```

```python
# Removing trailing and leading spaces from data
dataset = dataset.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

4. An amount more than
$5,000 should not be split
between two or more swipes
of the card by the same
person.

```python
# filter out the columns we want
filter_data = o_data[["CardholderLastName","CardholderFirstInitial", "Vendor", "TransactionDate", "Amount"]]
```

```python
# keep those with repeated entries
# in other words, transactions for two or more swipes
repeated_entries = filter_data[filter_data.duplicated(subset=["CardholderLastName","CardholderFirstInitial", "Vendor", "TransactionDate"], keep=False)]
repeated_entries
```

```python
# sum up for each same-day splitted entries
repeated_daily_entries=repeated_entries.groupby(["CardholderLastName","CardholderFirstInitial", "Vendor", "TransactionDate"])["Amount"].sum().reset_index(name ='DailyAmount')
```

```python
# Filter those with >5000 daily limits; sort by transaction dates
violate_with_split = repeated_daily_entries[repeated_daily_entries["DailyAmount"]>5000].sort_values(by="TransactionDate", ascending=True)
violate_with_split
```

5. Continued from 4. Count how often each individual purchaser did not follow the
previous restriction. Sort this in descending order by the count

```python
# count the number of violations for each purchaser
count_split_violation = violate_with_split.groupby(["CardholderLastName","CardholderFirstInitial"])["DailyAmount"].count().reset_index(name ='Violations')
```

```python
# sort with ascending order
count_split_violation.sort_values(by="Violations", ascending=False)
```

6. Purchases should not be
split between two or more
cardholders.

```python
filter_data.head() #contains the columns we want
```

```python
 # contains same name, same vendor, same date transactions
repeated_entries
```

```python
# contains all same vendor, same date transactions
repeated_vendors = filter_data[filter_data.duplicated(subset=["Vendor", "TransactionDate"], keep=False)]   
repeated_vendors
```

```python
# A split purchasing case is either splitted by (1) multiple swipes by one cardholder or (2) different cardholders
# Subtract (1) from all split purchasing orders, we expect to get (2)
merged = pd.merge(repeated_entries[["Vendor", "TransactionDate", "Amount"]], 
                  repeated_vendors[["Vendor", "TransactionDate", "Amount"]],
                    on=["Vendor", "TransactionDate"], how='outer',  suffixes=['_1', '_2'], indicator=True)
vendor_diff_cardholders = merged.loc[merged['_merge'] == 'right_only', ["Vendor", "TransactionDate", "Amount_2"]].rename(columns={'Amount_2': 'Amount'})
vendor_diff_cardholders
```

```python
# sum up for each same-day vendor amount with different purchasers
vendor_diff_card_sum = vendor_diff_cardholders.groupby(["Vendor", "TransactionDate"])["Amount"].sum().reset_index(name ='VendorDailyAmount')
vendor_diff_card_sum
```

```python
violate_with_diff_card = vendor_diff_card_sum[vendor_diff_card_sum["VendorDailyAmount"]>5000].sort_values(by="TransactionDate", ascending=True)
violate_with_diff_card
```

```python
# count number of violations
len(violate_with_diff_card)
```

7. Transactions are prohibited
for expenses for food and
mileage while traveling. A
per diem for food expenses
and mileage may be claimed
using a travel voucher.

```python
# Extract traveling dates of all purchasers
travel_dates = o_data[o_data["MerchantCategoryCode(MCC)"].str.contains('|'.join(['hotel', 'motel', 'resort', 'inn']))]
travel_dates = travel_dates[["CardholderLastName","CardholderFirstInitial", "TransactionDate"]].drop_duplicates()
travel_dates.sort_values(by="TransactionDate")
```

```python
# Extract all food related transactions
food_transactions = o_data[o_data["MerchantCategoryCode(MCC)"].str.contains('|'.join(['food', 'restaurant']))]
food_transactions.head()
```

```python
# Filter food transactions on those days with travelling
join = food_transactions.merge(travel_dates, how="inner", on=["CardholderLastName","CardholderFirstInitial","TransactionDate"])
food_violations = join[["Amount", "CardholderLastName", "CardholderFirstInitial", "Description", "Vendor", "TransactionDate", "PostedDate", "MerchantCategoryCode(MCC)"]
     ].sort_values(by=["CardholderLastName", "CardholderFirstInitial", "Amount"], ascending=[True, True,True])
food_violations
```

```python
# go one step further, count how many people have violated
len(food_violations.groupby(["CardholderLastName", "CardholderFirstInitial"])["Amount"].count())
```
